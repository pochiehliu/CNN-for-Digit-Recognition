# Exploratory Data Analysis and Visualization

The hand writting dataset is download from "The MNIST database" (http://yann.lecun.com/exdb/mnist/). Apply mnist package to import the data. Finally use matplotlib to check what the data looks like.

```python
import os
from mnist import MNIST
import numpy as np
from matplotlib import pyplot as plt
```

```python
mnidata = MNIST(os.getcwd()+'/Data')
```

```python
train_images, train_labels = mnidata.load_training()
test_images, test_labels = mnidata.load_testing()
```

```python
def squareFormat(x):
    square = np.array(x)
    square.resize(28,28)
    return square
    
def showImg(x,label):
    plt.imshow(x, cmap = 'gray')
    plt.title('Label: '+ str(label))
    plt.xticks([])
    plt.yticks([])
```

```python
fig = plt.figure(figsize = (15,5))
_ = fig.suptitle('First 10 data in training set')
for i in range(1,11):
    plt.subplot(2,5,i)
    showImg(squareFormat(train_images[i-1]), train_labels[i-1])
```

```python
fig = plt.figure(figsize = (15,5))
_ = fig.suptitle('First 10 data in testing set')
for i in range(1,11):
    plt.subplot(2,5,i)
    showImg(squareFormat(test_images[i-1]), test_labels[i-1])
```
